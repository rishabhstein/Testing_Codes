```python
import numpy as np;
import matplotlib.pyplot as plt


def generate_pt():
    a = 2 *(np.random.random_sample(1,)) - 1;
    return a;

def plot_pt(pt_x, pt_y):
    plt.plot(pt_x, pt_y, 'g.');
    plt.xlabel('X'); plt.ylabel('Y');
    plt.title("Random point generation")

def plot_circle(x,y,r,N):
    tmp = 0;
    for i in range(N):
        if (x[i]**2 + y[i]**2 <= r**2):
            plt.plot(x[i],y[i], 'r.')
            tmp += 1;
    return tmp;

```

```python
a = -1; b = 1; N = 2000;
x = []; y = [];
r = 1.;

for i in range(N):
    x.extend(generate_pt()); 
    y.extend(generate_pt());

plt.figure(figsize = [10,10])    
plot_pt(x,y)
M = plot_circle(x,y,r,N)


print('No. of points in the circle is: %d' %(M));
print('Value of Pi is OR Area of the circle is: %f' %(4*(M/N)));
```
